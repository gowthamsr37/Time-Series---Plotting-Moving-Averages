# Time Series - Plotting Moving Averages

```python
import pandas as pd 
import pandas_datareader as pdr
from datetime import datetime
```

```python
df_tesla = pdr.get_data_yahoo('TSLA')
df_tesla.head()
```

# Simple Moving Average


```python
##Smoothening the curve

df_tesla['Open'].plot(figsize = (16,6))
```

```python
df_tesla['Open:10 days rolling'] = df_tesla['Open'].rolling(window = 10 , min_periods=1).mean()

df_tesla[['Open' , 'Open:10 days rolling']].plot(figsize = (16,6), xlim = ['2020-01-01' , '2021-01-01'])
```

```python
df_tesla['Open:30 days rolling'] = df_tesla['Open'].rolling(window = 30 , min_periods=1).mean()

df_tesla[['Open' , 'Open:30 days rolling']].plot(figsize = (16,6), xlim = ['2020-01-01' , '2021-01-01'])
```

```python
df_tesla['Open:50 days rolling'] = df_tesla['Open'].rolling(window = 50 , min_periods=1).mean()

df_tesla[['Open' , 'Open:50 days rolling']].plot(figsize = (16,6), xlim = ['2020-01-01' , '2021-01-01'])
```

```python
df_tesla[['Open' , 'Open:10 days rolling', 'Open:30 days rolling' , 'Open:50 days rolling']].plot(figsize = (16,6), xlim = ['2020-01-01' , '2021-01-01'] , ylim = [0,800])
```

Disadvantages: We are giving equal importance to all the data

# Cumulative Moving Average


```python
df_tesla['Open'].expanding().mean().plot(figsize = (10,5))
```

# Exponential Moving Average
Focus will be with the recent data


```python
df_tesla['EMA_0.1'] = df_tesla["Open"].ewm(alpha = 0.1 , adjust = False ).mean()
df_tesla[['Open' , 'EMA_0.1']].plot(xlim = ['2020-01-01' , '2021-01-01'] , figsize = (15,6))
```

```python
df_tesla['EMA_0.3'] = df_tesla["Open"].ewm(alpha = 0.3 , adjust = False ).mean()
df_tesla[['Open' ,'EMA_0.1', 'EMA_0.3']].plot(xlim = ['2020-01-01' , '2021-01-01'] , figsize = (15,6))
```

## Exponential Moving Average


```python
df_tesla['EMA_5_days'] = df_tesla["Open"].ewm(span =5 ).mean()
df_tesla[['Open' ,'EMA_5_days']].plot( figsize = (15,6))
```

```python
df_tesla[['Open' ,'EMA_5_days' , 'EMA_0.1']].plot( figsize = (15,6))
```
